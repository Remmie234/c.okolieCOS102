```python
import numpy as np

def solution(roots):
    for r in roots:
        if r.imag == 0:
            print(f"Real Root: {r.real:.2f}")
        else:
            sign = "+" if r.imag >= 0 else "-"
            print(f"Complex Root: {r.real:.2f} {sign} {abs(r.imag):.2f}i")
    print("--------------------------------------------------")
    print("\n")
            
while True:
    print("--- Project 2: Calculator ---")
    print("a) Quadratic Equation\n"
          "b) Cubic Equation\n"
          "c) Quartic Equation\n"
          "q) quit\n")
    choice = input("Select an option a, b, c, or q: ").lower()

    #Terminate
    if choice == 'q':
        print("Exiting...")
        break
        
    if choice == 'a':
        a = float(input("Enter the value of a: "))
        b = float(input("Enter the value of b: "))
        c = float(input("Enter the value of c: "))
        roots = np.roots([a, b, c])
        solution(roots)
        
    elif choice == 'b':
        #Using Cardano's Method
        a = float(input("Enter the value of a: "))
        b = float(input("Enter the value of b: "))
        c = float(input("Enter the value of c: "))
        d = float(input("Enter the value of d: "))
        roots = np.roots([a, b, c, d])
        solution(roots)
        
    elif choice == 'c':
        a = float(input("Enter the value of a: "))
        b = float(input("Enter the value of b: "))
        c = float(input("Enter the value of c: "))
        d = float(input("Enter the value of d: "))
        e = float(input("Enter the value of e: "))
        roots = np.roots([a, b, c, d, e])
        solution(roots)

    else:
        print("Invalid selection. Please run the program again and choose a, b, c, or q.\n")
```
