```python
while True:
    print("\n--- Annual Tax Revenue Calculator (ATR) ---")
    name = input("Kindly enter your name: ").strip()
    
    # Number Validation
    try:
        age = int(input(f"{name}, kindly enter your age: "))
        exp = int(input(f"{name}, kindly enter your years of experience: "))
    except ValueError:
        print("Error: Please enter whole numbers for age and experience.")
        continue

    # Logical Conditioning
    if age < 0 or age > 120 or exp < 0 or exp > 100:
        print("Invalid age or experience values. Please restart.")
        continue
    
    if exp > (age - 14): # Assuming work starts at earliest age 14
        print(f"Error: Experience ({exp}) is impossible for an age of {age}.")
        continue

    # Logic
    if age >= 55 and exp > 25:
        ATR = 5_600_000
    elif age >= 45 and exp > 20:
        ATR = 4_480_000
    elif age >= 35 and exp >= 10:
        ATR = 1_500_000
    else:
        ATR = 550_000

    # Receipt
    print("-" * 37)
    print(f"Name: {name}\nAge: {age}\nExperience: {exp}")
    print(f"Annual Tax Revenue: {ATR:,.0f}₦")
    print("-" * 37)

    # Stronger Iteration Logic
    while True:
        choice = input(f"{name}, do you wish to end calculator (y/n): ").lower()
        if choice in ['y', 'n']:
            break #For question loop
        print("Invalid selection. Please type 'y' or 'n'.")

    if choice == 'y':
        print("Exiting...")
        break #For entire calculator
```
