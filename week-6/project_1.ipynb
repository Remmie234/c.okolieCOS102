```python
import tkinter as tk
from tkinter import messagebox as msgbox
import random

class EmployeeSystem:
    def __init__(self, root):
        self.root = root
        self.root.title("Mrs. Jane's Logistics System")
        self.root.geometry("400x300")
        
        self.employees = [
            "Mary Evans", "Eyo Ishan", "Durojaiye Dare", "Adams Ali", "Andrew Ugwu",
            "Stella Mankinde", "Jane Akibo", "Ago James", "Michell Taiwo", 
            "Abraham Jones", "Nicole Anide", "Kosi Korso", "Adele Martins", 
            "Emmanuel Ojo", "Ajayi Fatima"
        ]
        self.tasks = ["Loading", "Transporting", "Reviewing Orders", "Customer Service", "Delivering Items"]

        #GUI Elements
        tk.Label(root, text="Employee Verification System", font=("Helvetica", 14, "bold")).pack(pady=20)
        
        tk.Label(root, text="Enter Full Name:").pack()
        self.name_entry = tk.Entry(root, width=30)
        self.name_entry.pack(pady=10)

        self.verify_btn = tk.Button(root, text="Login / Verify", command=self.check_employee, bg="green", fg="white")
        self.verify_btn.pack(pady=20)

    def check_employee(self):
        user_input = self.name_entry.get().strip()
        if user_input in self.employees:
            self.take_attendance(user_input)
        else:
            self.refuse_access()

    def take_attendance(self, name):
        self.assign_task(name)

    #To randomize
    def assign_task(self, name):
        task_index = random.randint(0, len(self.tasks) - 1)
        assigned_task = self.tasks[task_index]
        msgbox.showinfo("Success", f"Welcome, {name}!\nAttendance marked.\n\nTask: {assigned_task}")

    def refuse_access(self):
        msgbox.showwarning("Access Denied", "Sorry, your name was not found in our records.")

# Initialize and run app
root = tk.Tk()
app = EmployeeSystem(root)
root.mainloop()
```
