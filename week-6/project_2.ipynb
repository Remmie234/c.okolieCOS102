```python
import tkinter as tk
from tkinter import messagebox as msgbox

class DeliveryService:
    def __init__(self, root):
        self.root = root
        self.root.title("Delivery Price Calculator")
        self.root.geometry("350x300")

        #UI
        tk.Label(root, text="Package Weight (kg):").pack(pady=5)
        self.weight_entry = tk.Entry(root)
        self.weight_entry.pack(pady=5)

        tk.Label(root, text="Select Location:").pack(pady=5)
        #I use aStringVar to keep track of the dropdown selection
        self.location_var = tk.StringVar(root)
        self.location_var.set("PAU") #It's default
        
        self.location_menu = tk.OptionMenu(root, self.location_var, "PAU", "Epe")
        self.location_menu.pack(pady=5)

        tk.Button(root, text="Calculate Price", command=self.calculate_price, bg="blue", fg="white").pack(pady=20)

    def calculate_price(self):
        try:
            #Grab data
            weight = float(self.weight_entry.get())
            location = self.location_var.get()
            price = 0

            #PAU's Logic
            if location == "PAU":
                if weight >= 10:
                    price = 2000
                else:
                    price = 1500
            
            #Epe's Logic
            elif location == "Epe":
                if weight >= 10:
                    price = 5000
                else:
                    price = 4000

            msgbox.showinfo("Quote", f"The delivery charge to {location} is: N{price}")
            
        except ValueError:
            msgbox.showerror("Error", "Please enter a valid number for the weight.")

#Run app
root = tk.Tk()
app = DeliveryService(root)
root.mainloop()
```
