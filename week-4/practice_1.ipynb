```python
score = 50 # Global variable

def update_score():
    global score # Tells Python: "Use the score from outside"
    score = score + 10
    print("Inside function:", score)

def check_score():
    score = 100 # No global keyword, so this creates a NEW local variable
    print("Inside check:", score)

update_score()
# Output: Inside function: 60

check_score()
# Output: Inside check: 100

print("Final Global Score:", score)

# Output: Final Global Score: 60 (Modified by update_score, but NOT by check_score)
```
