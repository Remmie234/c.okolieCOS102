```python
while True:
    print("--- Project 1: Calculator ---")
    print("a) Simple Interest\n"
          "b) Compound Interest\n"
          "c) Annuity Plan\n"
          "q) quit\n")
    choice = input("Select an option (a, b, c or q): ").lower()
        
    # Terminate
    if choice == 'q':
        print("Exiting...")
        break 
        
    if choice == 'a':
        #Designing the Simple Interest
        P = float(input("Input the Principal (P): "))
        r = float(input("Input the annual rate in % (r): "))
        t = float(input("Input the years (t): "))
        #Amount
        A = P * (1 + r/100 * t)
        print(f"The Total Amount is {A:.2f}")
        #The Simple Interest
        I = A - P
        print(f"The Simple Interest is {I:.2f}\n")
        print("----------------------------------------------------")


    elif choice in ['b', 'c']:
        # Efficiently asking for "n, r, and t" if they Compound or Annuity is chosen
        r = float(input("Annual rate % (r): "))
        n = float(input("Times per year (n): "))
        t = float(input("Years (t): "))

        #Simply assigning variables and converting to decimal for goodlook
        rate_per_period = ((r / 100) / n)
        total_periods = (n * t)
    
        if choice == 'b':
            #Designing Compound Interest
            P = float(input("Principal (P): "))
            #Amount
            A = P * (1 + rate_per_period) ** total_periods
            print(f"The Total Amount is {A:.2f}")
            #The Compound interest
            I = A - P
            print(f"The Compound Interest is {I:.2f}\n")
            print("----------------------------------------------------")

    
        else:
            #Designing the Annuity Plan
            PMT = float(input("Input the periodic payment (PMT): ")) 
            # The Annuity Formula
            A = PMT * (((1 + rate_per_period) ** total_periods - 1) / rate_per_period)
            print(f"The Annuity Plan is {A:.2f}\n")
            print("----------------------------------------------------")
    
    else:
        print("Invalid selection. Please run the program again and choose a, b, c or q.")
```
