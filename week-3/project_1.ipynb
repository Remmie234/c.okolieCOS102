```python
import pandas as pd
import tkinter as tk
from tkinter import messagebox as msgbox

#Importing CSV file
file = pd.read_csv('GIG-logistics(in).csv')

#Input & Validity
def auth():
    name_input = employee_entry.get().strip().lower()
    dept_input = department_entry.get().strip().lower()

    mask = (file['Name'].str.strip().str.lower() == name_input) & \
           (file['Department'].str.strip().str.lower() == dept_input)
    
    results = file[mask]

    # Validating logic
    if not results.empty:
        #Their name
        real_name = results['Name'].values[0]
        dept_name = results['Department'].values[0]
        #Other members
        others = file[file['Department'] == dept_name]
        
        msgbox.showinfo("Success", f"Welcome, {real_name}!\n\nDepartment Peers:\n{others['Name'].to_string(index=False)}")
        print(others)

        #Close view
        answer = msgbox.askyesno("Confirmation", "Would you like to close the view?")
        if answer: 
            root.destroy()
    else:
        msgbox.showerror("Denied", "Employee does not exist in our records.")
        
#Create main window
root = tk.Tk()
root.title("Auth Form")
root.geometry("500x200")

#Employee name and label
employee_name = tk.Label(root, text="Employee name: ")
employee_name.pack()
employee_entry = tk.Entry(root)
employee_entry.pack()

#Department and label
department_name = tk.Label(root, text="Department")
department_name.pack()
department_entry = tk.Entry(root)
department_entry.pack()

#Submit button and label
submit_button = tk.Button(root, text="Submit", command=auth)
submit_button.pack()
    
#Run main window
root.mainloop()
```
