```python
import tkinter as tk
from tkinter import messagebox as msgbox

class Zenith:
    def __init__(self, employee_name):
        self.employee_name = employee_name

    def unique_services(self):
        return []

    def mutual_services(self):
        return ["General Banking Support"]

class RetailBanking(Zenith):
    def unique_services(self):
        return [
            "Retirement and education accounts",
            "Loans and mortgages",
            "Checking and saving"
        ]

    def mutual_services(self):
        return ["Lines of credit", "Investment management", "Insurance"]

class GlobalBanking(Zenith):
    def unique_services(self):
        return [
            "Multi-currency management services",
            "Foreign currency accounts",
            "Foreign currency credit cards",
            "Transborder advisory services",
            "Liquidity management"
        ]

class CommercialBanking(Zenith):
    def unique_services(self):
        return ["Advisory services"]

    def mutual_services(self):
        return ["Lines of credit", "Investment management", "Insurance"]

#This is my GUI Logic

class ZenithGUI:
    def __init__(self, window):
        self.window = window
        self.window.title("Zenith Bank Division Portal")
        self.window.geometry("450x500")

        #Employee Name
        tk.Label(window, text="Employee Name:", font=("Arial", 10, "bold")).pack(pady=5)
        self.ent_name = tk.Entry(window, width=30)
        self.ent_name.pack(pady=5)

        #Division Selection
        tk.Label(window, text="Select Division:", font=("Arial", 10, "bold")).pack(pady=5)
        self.selected_division = tk.StringVar(window)
        self.selected_division.set("Retail Banking") 
        divisions = ["Retail Banking", "Global Banking", "Commercial Banking"]
        self.drop_menu = tk.OptionMenu(window, self.selected_division, *divisions)
        self.drop_menu.pack(pady=5)

        #Display Button
        self.btn_show = tk.Button(window, text="Display Services", command=self.process_request, bg="#E11A22", fg="white")
        self.btn_show.pack(pady=20)

        #Output Area
        self.txt_display = tk.Text(window, height=12, width=50, state='disabled', bg="#f4f4f4")
        self.txt_display.pack(pady=10)

    def process_request(self):
        name = self.ent_name.get()
        div = self.selected_division.get()

        if not name:
            msgbox.showwarning("Input Required", "Please enter the employee's name.")
            return

        #I'm instantiating with respect to the selection
        if div == "Retail Banking":
            emp_obj = RetailBanking(name)
        elif div == "Global Banking":
            emp_obj = GlobalBanking(name)
        else:
            emp_obj = CommercialBanking(name)

        self.show_output(emp_obj, div)

    def show_output(self, obj, div_name):
        #For accessing inherited and polymorphic methods
        services = obj.unique_services() + obj.mutual_services()
        
        self.txt_display.config(state='normal')
        self.txt_display.delete('1.0', tk.END)
        
        result = f"Employee: {obj.employee_name}\n"
        result += f"Division: {div_name}\n"
        result += "="*30 + "\n"
        result += "SERVICES RENDERED:\n"
        for item in services:
            result += f"• {item}\n"
            
        self.txt_display.insert(tk.END, result)
        self.txt_display.config(state='disabled')

    root = tk.Tk()
    app = ZenithGUI(root)
    root.mainloop()
```
